```python
# Importing Necssary Pacakages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('melbourne_house_price.csv')
df.head()
```

```python
df.describe()
```

```python
df.nunique()
```

```python
df.shape
```

```python
# Now we will remove the unwanted features that are not needed
New_features = ['Suburb', 'Rooms', 'Type', 'Method', 'Regionname', 'Propertycount', 'Distance',
               'CouncilArea', 'Bedroom2','Car','BuildingArea','Price','Landsize','SellerG','Bathroom']
df = df[New_features]
df.head()
```

```python
df.shape
```

```python
# Now we will handle the missing values or null values
df.isna().sum()
```

```python
# We will fill the null values of the ['Bedroom2', 'Car','Propertycount','Distance','Bathroom'] to 0
Extra_features = ['Bedroom2', 'Car','Propertycount','Distance','Bathroom']
df[Extra_features] = df[Extra_features].fillna(0)
df.head()
```

```python
df.isna().sum()
```

```python
# Now filling the null values of the BuildingArea and Landsize as mean value
df['BuildingArea'] = df["BuildingArea"].fillna(df.BuildingArea.mean())
df['Landsize'] = df["Landsize"].fillna(df.Landsize.mean())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.head()
```

**We will create dummi variables of categorical values**

```python
df = pd.get_dummies(df,drop_first=True)
df.head()
```

```python
#  Now remove outliers using z score. Use z score of 3 as your thresold
df['zscore'] = (df.Price-df.Price.mean())/df.Price.std()
df.head()
```

```python
outliers_z = df[(df.zscore < -3) | (df.zscore > 3)]
outliers_z.shape
```

```python
new_df = df[(df.zscore>-3)&(df.zscore<3)]
new_df.shape
```

```python
new_df.describe()
```

**Now we are ready to bulild our model**

```python
X = df.drop(['Price'],axis = 'columns')
y = df.Price
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state=0) 
print(x_train.shape,y_train.shape,x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

linear_reg = LinearRegression()

linear_reg.fit(x_train,y_train)
```

```python
coef = linear_reg.coef_
```

```python
intercept = linear_reg.intercept_
intercept
```

```python
bias = linear_reg.score(x_train,y_train)
bias
```

```python
variance = linear_reg.score(x_test,y_test)
variance
```

```python
plt.scatter(linear_reg.predict(x_test),linear_reg.predict(x_test)-y_test, color="green", s=12, label="Test data")
 
plt.scatter(linear_reg.predict(x_train),linear_reg.predict(x_train)-y_train, color="orange", s=12, label="Train data")

plt.hlines(y=0, xmin=50,xmax=180,linewidth=2)
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics

predictions = linear_reg.predict(x_test)  
mse = mean_squared_error(y_test, predictions) 
rmse = np.sqrt(mse)

print("RMSE on Predicted output value : ", rmse)
print("MAE on Predicted output value : ", metrics.mean_absolute_error(y_test,predictions))
print("MSE on Predicted output value : ", metrics.mean_squared_error(y_test,predictions))
```

```python
from sklearn.metrics import accuracy_score, r2_score
print("r2_Score : ", r2_score(y_test,predictions))
```
